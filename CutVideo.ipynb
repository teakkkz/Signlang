```python
import cv2
print(cv2.__version__)
vidcap = cv2.VideoCapture('videos/Sun_2.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("sun(2)%d.jpg" % count, image)     # save frame as JPEG file
  success,image = vidcap.read()
  print ('Read a new frame: ', success)
  count += 1
```
